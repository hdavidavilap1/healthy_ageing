```python
pd_diabetes = pd.read_csv('/home/hdap/Documents/bkDavid/maestria_unir/practicas/data_david/diabetes/total_diabetes.csv', sep = '|')
pd_depresion = pd.read_csv('/home/hdap/Documents/bkDavid/maestria_unir/practicas/data_david/depression/total_depression.csv', sep = '|')
pd_cardio = pd.read_csv('/home/hdap/Documents/bkDavid/maestria_unir/practicas/data_david/cardiovascular/total_cardiovascular.csv', sep = '|')
pd_respiratory = pd.read_csv('/home/hdap/Documents/bkDavid/maestria_unir/practicas/data_david/respiratory/total_respiratory.csv', sep = '|')
```

```python
word = 'DR2LANG'

print([c for c in pd_diabetes.columns if word in c])
print([c for c in pd_depresion.columns if word in c])
print([c for c in pd_cardio.columns if word in c])
print([c for c in pd_respiratory.columns if word in c])
print([c for c in pd_final.columns if word in c])
```

```python
pd_final = pd_final.merge(pd_diabetes, how = 'left', on = 'SEQN')
pd_final = pd_final.merge(pd_depresion, how = 'left', on = 'SEQN')
pd_final = pd_final.merge(pd_cardio, how = 'left', on = 'SEQN')
pd_final = pd_final.merge(pd_respiratory, how = 'left', on = 'SEQN')

pd_final.shape
```

```python
target = 'enfermedad'
column_edad = 'RIDAGEYR'

threshold = [50, 60, 70, 80]

[sum(np.where(pd_final[column_edad] >= thr, 1, 0)) for thr in threshold]
```

## Creacion de la variable dependiente

```python
#Este es e threshold de la edad
thr = 50
```

```python
pd_final_edad = pd_final[pd_final[column_edad] >= thr]
pd_final_edad.shape
```

```python
# Se crea una variable que resume si la persona tiene algún tipo de enfermedad
```

```python
pd_final_edad.loc[:,'enfermedad'] = np.where((pd_final_edad.enfermedad_diabetes == 1) | (pd_final_edad.enfermedad_depresion == 1) | (pd_final_edad.enfermedad_cardiovascular == 1) | (pd_final_edad.enfermedad_respiratoria == 1), 1, 0)
```

```python
pd_final_edad.shape
```

```python
pd_final_edad['enfermedad_diabetes'] = pd_final_edad['enfermedad_diabetes'].fillna(0)
pd_final_edad['enfermedad_depresion'] = pd_final_edad['enfermedad_depresion'].fillna(0)
pd_final_edad['enfermedad_cardiovascular'] = pd_final_edad['enfermedad_cardiovascular'].fillna(0)
pd_final_edad['enfermedad_respiratoria'] = pd_final_edad['enfermedad_respiratoria'].fillna(0)
```

```python
print(pd_final_edad.enfermedad_diabetes.sum())
print(pd_final_edad.enfermedad_depresion.sum())
print(pd_final_edad.enfermedad_cardiovascular.sum())
print(pd_final_edad.enfermedad_respiratoria.sum())
print(pd_final_edad.enfermedad.sum())
```

## Manejo de Datos Faltantes

```python
total_filas = pd_final_edad.shape[0]
max_na_col = 0.2*total_filas
```

```python
pd_describe = pd_final_edad.describe().T
pd_describe.shape
```

```python
columnas_utilizar = pd_describe[pd_describe['count'] > max_na_col].index
```

```python
pd_final_edad = pd_final_edad[columnas_utilizar]
```

```python
## imputa la media

for c in columnas_utilizar:
    media = pd_final_edad[c].mean()
    pd_final_edad[c] = pd_final_edad[c].fillna(media) 
```

```python
pd_final_edad.enfermedad.value_counts()
```

```python
pd_train, pd_test = ms.train_test_split(pd_final_edad,test_size = 0.3, stratify = pd_final_edad[target])
```

```python
[col for col in pd_final_edad.columns if 'enfermedad' in col]
```
