```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import traceback
```

```python
leer = True
```

```python

if leer:
    anos = ['2003-2004', '2005-2006', '2007-2008', '2009-2010', '2011-2012', '2013-2014','2015-2016', '2017-2018']
    
    extension = ['DIQ_C','DIQ_D','DIQ_E','DIQ_F','DIQ_G','DIQ_H','DIQ_I','DIQ_J','DIQ_K']
    
    archivos = []
    
    for f,j in zip(anos,extension):
        try:
            print(f)
            print('-----------------------------')
            path = 'https://wwwn.cdc.gov/Nchs/Nhanes/{}/{}.XPT'.format(f, j)
            print(path)
            archivo = pd.read_sas(path, format='xport', encoding='utf-8')
            
            if f in ['2003-2004']:
                archivo = archivo[(archivo.DIQ010==1)|(archivo.DIQ050==1)|(archivo.DIQ070==1)|(archivo.DIQ080==1)|(archivo.DIQ090==1)|(archivo.DIQ100==1)|(archivo.DIQ120==1)|(archivo.DIQ140==1)]
            elif f in ['2005-2006', '2007-2008']:
                archivo = archivo[(archivo.DIQ010==1)|(archivo.DIQ160==1)|(archivo.DIQ050==1)|(archivo.DIQ240==1)|(archivo.DIQ080==1)]
            else: 
                archivo = archivo[(archivo.DIQ010==1)|(archivo.DIQ160==1)|(archivo.DIQ050==1)|(archivo.DIQ070==1)|(archivo.DIQ240==1)|(archivo.DIQ080==1)] 
            print(archivo.shape)
            archivo = archivo[['SEQN']]
            archivo[['enfermedad_diabetes']] = 1
            archivos.append(archivo)
        except Exception as exc:
            print(path)
            print(traceback.format_exc())
            print(exc)
            print('--------------------')
            pass
    
    pd_ = pd.concat(archivos, sort='False', ignore_index='True')
    pd_ = pd_.drop_duplicates()
    pd_.to_csv('./total_diabetes.csv', header = True, sep = '|', index = False)
    print(pd_.shape)
```
